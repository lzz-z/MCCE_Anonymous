```python
from rdkit import Chem
from rdkit.Chem import AllChem, Descriptors
from pyscf import gto, scf

def smiles_to_mol(smiles: str):
    """生成RDKit分子对象和3D坐标"""
    mol = Chem.MolFromSmiles(smiles)
    mol = Chem.AddHs(mol)
    AllChem.EmbedMolecule(mol, AllChem.ETKDG())
    AllChem.MMFFOptimizeMolecule(mol)
    return mol

def compute_properties(smiles: str):
    mol = smiles_to_mol(smiles)

    # 1. 分子量
    mw = Descriptors.MolWt(mol)

    # 2. 导出几何结构
    atoms = []
    conf = mol.GetConformer()
    for atom in mol.GetAtoms():
        pos = conf.GetAtomPosition(atom.GetIdx())
        atoms.append(f"{atom.GetSymbol()} {pos.x} {pos.y} {pos.z}")
    atom_str = "; ".join(atoms)

    # 3. PySCF计算 HOMO–LUMO gap 和 dipole
    mol_pyscf = gto.Mole()
    mol_pyscf.build(atom=atom_str, basis="sto-3g", unit="Angstrom")
    mf = scf.RHF(mol_pyscf).run()

    mo_energy = mf.mo_energy
    n_occ = mol_pyscf.nelectron // 2
    homo = max(mo_energy[:n_occ])
    lumo = min(mo_energy[n_occ:])
    gap = lumo - homo  # au

    # 偶极矩 (a.u. → Debye, 1 a.u. = 2.541746 D)
    dipole_au = mf.dip_moment(unit='au')
    dipole = (sum(x**2 for x in dipole_au) ** 0.5) * 2.541746

    return {
        "molecular_weight": mw,
        "homo_lumo_gap_au": gap,
        "dipole_debye": dipole
    }

if __name__ == "__main__":
    smiles = "CC(=O)OC1=CC=CC=C1C(=O)O"  # 这里换成你要测试的SMILES
    props = compute_properties(smiles)
    print(props)
```

```python
import json
import os
import numpy as np


def top_auc(buffer, top_n, finish, freq_log, max_oracle_calls):
    sum = 0
    prev = 0
    called = 0
    ordered_results = list(sorted(buffer, key=lambda kv: kv[1], reverse=False))
    for idx in range(freq_log, min(len(buffer), max_oracle_calls), freq_log):
        temp_result = ordered_results[:idx]
        temp_result = list(sorted(temp_result, key=lambda kv: kv[0].total, reverse=True))[:top_n]
        top_n_now = np.mean([item[0].total for item in temp_result])
        sum += freq_log * (top_n_now + prev) / 2
        prev = top_n_now
        called = idx
    temp_result = list(sorted(ordered_results, key=lambda kv: kv[0].total, reverse=True))[:top_n]
    top_n_now = np.mean([item[0].total for item in temp_result])
    sum += (len(buffer) - called) * (top_n_now + prev) / 2
    if finish and len(buffer) < max_oracle_calls:
        sum += (max_oracle_calls - len(buffer)) * top_n_now
    return sum / max_oracle_calls

def cal_avg(root,file_num,metric,start_seed = 42):
    results = None
    for i in range(file_num):
        filename = root.format(start_seed+i)
        with open(os.path.join(root,filename),'r') as f:
            file = json.load(f)
        names = file['results'][0].keys()
        if results == None:
            results = {name:[] for name in names}
        
        for key in  results.keys():
            results[key].append(file['results'][-1][key])

    #print(results['avg_top1'])
    mean = np.mean(results[metric])
    std = np.std(results[metric])
    print(f'{metric}: {mean:.3f} ± {std:.3f}')
    return round(float(mean),4),round(float(std),4)
gemini_f10_mean = []
gemini_f10_std = []
for i in range(1,7):
#for i in [3,7,9]:
    #root = f'/root/nian/mollm_results/koff/zgca,gemini-2.5-flash-nothinking/results/sa_drd2_qed_gsk3b_jnk3_0919_noexp{i}' + 'off_{}.json'
    #root = f'/root/nian/mollm_results/koff/zgca,gpt-4o-2024-05-13/results/sa_drd2_qed_gsk3b_jnk3_0919_noexp{i}' + 'off_{}.json'
    #root = f'/root/nian/mollm_results/koff/zgca,gpt-4o-2024-05-13/results/jnk3_0919_noexp{i}' + 'off_{}.json'
    #root = f'/root/nian/mollm_results/koff/zgca,gemini-2.5-flash-nothinking/results/jnk3_0919_noexp{i}' + 'off_{}.json'
    #root = f'/root/nian/mollm_results/exp/zgca,gpt-4o-2024-05-13/results/sa_drd2_qed_gsk3b_jnk3_0919_exp0.{i}'+'_{}.json'
    root = f'/root/nian/mollm_results/exp/zgca,gemini-2.5-flash-nothinking/results/sa_drd2_qed_gsk3b_jnk3_0919_exp0.{i}'+'_{}.json'
    #mean,std = cal_avg(root,file_num=3,metric='hypervolume')
    mean,std = cal_avg(root,file_num=3,metric='div') # avg_top10
    gemini_f10_mean.append(mean)
    gemini_f10_std.append(std)
```

```python
import requests
import time
import pandas as pd
import re
from rdkit import Chem
from rdkit.Chem import Descriptors, QED, rdMolDescriptors

# --------- LLM Query 接口 ----------
def proxy_chat(content):
    base_url = "http://35.220.164.252:3888/v1/chat/completions"
    api_key = "sk-FTse9VUA3G6wV7fTQRXHrvokcFJvDAJP6NoDhmb1Oy6ssBia"
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {api_key}"
    }
    data = {
        "model": 'gpt-5-minimal',
        "messages": [
            {"role": "user", "content": content}
        ]
    }
    while True:
        try:
            response = requests.post(base_url, headers=headers, json=data)
            response.raise_for_status()
            if response.status_code == 200:
                break
            else:
                print(f"Request failed {response.status_code}: {response.text}")
        except Exception as e:
            print(f'Exception {e}, retry in 20s')
            time.sleep(20)
    response = response.json()
    return response['choices'][0]['message']['content']


# --------- 提取生成的候选分子 ----------
def extract_smiles_from_string(text):
    pattern = r"<candidate>(.*?)</candidate>"
    smiles_list = re.findall(pattern, text, flags=re.DOTALL)
    return [s.strip() for s in smiles_list]


from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit import RDLogger
RDLogger.DisableLog('rdApp.*')
from pyscf import gto, scf
from tdc import Oracle

def smiles_to_xyz(smiles):
    """把 SMILES 转 xyz 坐标字符串"""
    mol = Chem.MolFromSmiles(smiles)
    mol = Chem.AddHs(mol)
    AllChem.EmbedMolecule(mol)
    AllChem.UFFOptimizeMolecule(mol)

    conf = mol.GetConformer()
    xyz = ""
    for i in range(mol.GetNumAtoms()):
        pos = conf.GetAtomPosition(i)
        sym = mol.GetAtomWithIdx(i).GetSymbol()
        xyz += f"{sym} {pos.x:.6f} {pos.y:.6f} {pos.z:.6f}\n"
    return xyz

def run_pyscf(smiles):
    """用 PySCF 计算 HOMO-LUMO gap (eV) 和偶极矩 (Debye)"""
    xyz = smiles_to_xyz(smiles)
    mol = gto.M(atom=xyz, basis='sto-3g', unit='Angstrom',verbose=0)
    mf = scf.RHF(mol).run()

    # 轨道能级 (Hartree → eV)
    mo_energies = mf.mo_energy
    homo = max([e for (e, occ) in zip(mo_energies, mf.mo_occ) if occ > 0])
    lumo = min([e for (e, occ) in zip(mo_energies, mf.mo_occ) if occ == 0])
    gap = (lumo - homo)/27.21 #* 27.2114  # 转 eV

    # 偶极矩 (Debye)
    dip = mf.dip_moment(unit='Debye',verbose=0)
    return gap, (sum(d**2 for d in dip)**0.5)

sa_scorer = Oracle(name = 'SA')
# --------- 分子目标函数计算 ----------
def calc_properties(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return None
        mw = Descriptors.MolWt(mol)
        sa = float(sa_scorer(smiles))
        homo_lumo_gap, dipole_moment = run_pyscf(smiles)
        return {
            "molecular_weight": mw,
            "sa_score": sa,
            "homo_lumo_gap": homo_lumo_gap,
            "dipole_moment": dipole_moment
        }
    except Exception as e:
        print("计算失败:", e)


# --------- 约束条件检查 ----------
def check_constraints(props, constraint):
    if props is None:
        return False
    if not (constraint["molecular_weight"][0] <= props["molecular_weight"] <= constraint["molecular_weight"][1]):
        return False
    if not (constraint["sa_score"][0] <= props["sa_score"] <= constraint["sa_score"][1]):
        return False
    if not (constraint["homo_lumo_gap"][0] <= props["homo_lumo_gap"] <= constraint["homo_lumo_gap"][1]):
        return False
    if not (constraint["dipole_moment"][0] <= props["dipole_moment"] <= constraint["dipole_moment"][1]):
        return False
    return True


# --------- 主流程 ----------
constraints = [
    {"constraint_id": "test_001", "molecular_weight": [200, 350], "sa_score": [1, 4], "homo_lumo_gap": [0.25, 0.35], "dipole_moment": [1.0, 3.0]},
{"constraint_id": "test_002", "molecular_weight": [350, 500], "sa_score": [3, 6], "homo_lumo_gap": [0.20, 0.30], "dipole_moment": [2.0, 5.0]},
{"constraint_id": "test_003", "molecular_weight": [150, 250], "sa_score": [1, 3.5], "homo_lumo_gap": [0.30, 0.40], "dipole_moment": [0.0, 1.5]},
{"constraint_id": "test_004", "molecular_weight": [250, 400], "sa_score": [2.5, 5.5], "homo_lumo_gap": [0.22, 0.32], "dipole_moment": [4.0, 7.0]},
{"constraint_id": "test_005", "molecular_weight": [280, 320], "sa_score": [2.0, 3.5], "homo_lumo_gap": [0.29, 0.31], "dipole_moment": [1.5, 2.5]}
]
import pandas as pd
zinc = pd.read_csv('/root/nian/zinc250k.csv')
zinc.loc[0,'smiles'].replace('\n','') # 如何取smiles
if "smiles" in zinc.columns:
    smiles_list = zinc["smiles"].dropna().tolist()
```

```python
from typing import List, Tuple, Dict, Any
from typing import Optional
import numpy as np
import pandas as pd
from scipy.optimize import minimize
def _compute_total_volume(solution_df: pd.DataFrame, combos_map: Dict[str, Dict[str, float]]) -> float:
    """用组合的 flow_rate_m3h * 时长小时，计算总输送量（m³）"""
    sol = solution_df.copy()
    st = pd.to_datetime(sol["start_time"])
    et = pd.to_datetime(sol["end_time"])
    hours = (et - st).dt.total_seconds() / 3600.0
    flows = sol["combination_id"].map(lambda cid: combos_map[cid]["flow"])
    return float((flows * hours).sum())

def snap_total_volume_to_exact(task_start_iso: str,
                               solution_df: pd.DataFrame,
                               combos_map: Dict[str, Dict[str, float]],
                               required_volume: float = 25000.0,
                               tol: float = 1e-6) -> pd.DataFrame:
    """
    仅调整最后一段 end_time，使总输送量精确逼近 required_volume。
    不跨天，不破坏连续性，不改变组合与切换次数。
    """
    if solution_df.empty:
        return solution_df

    sol = solution_df.copy()
    # 统一为 Timestamp 便于计算
    sol["start_time"] = pd.to_datetime(sol["start_time"])
    sol["end_time"]   = pd.to_datetime(sol["end_time"])

    # 计算当前体积
    current_volume = _compute_total_volume(sol, combos_map)
    delta = required_volume - current_volume
    if abs(delta) <= tol:
        # 已经足够接近
        # 转回 isoformat 字符串
        sol["start_time"] = sol["start_time"].dt.strftime("%Y-%m-%dT%H:%M:%S.%f").str.rstrip("0").str.rstrip(".")
        sol["end_time"]   = sol["end_time"].dt.strftime("%Y-%m-%dT%H:%M:%S.%f").str.rstrip("0").str.rstrip(".")
        return sol

    # 任务当天上界
    base_midnight = pd.to_datetime(task_start_iso)
    day_end = base_midnight.normalize() + pd.Timedelta(days=1)  # 当天24:00

    # 取最后一段
    last_idx = len(sol) - 1
    last_start = sol.loc[last_idx, "start_time"]
    last_end   = sol.loc[last_idx, "end_time"]
    last_cid   = sol.loc[last_idx, "combination_id"]

    if last_cid not in combos_map:
        # 数据异常，直接返回
        sol["start_time"] = sol["start_time"].dt.strftime("%Y-%m-%dT%H:%M:%S.%f").str.rstrip("0").str.rstrip(".")
        sol["end_time"]   = sol["end_time"].dt.strftime("%Y-%m-%dT%H:%M:%S.%f").str.rstrip("0").str.rstrip(".")
        return sol

    flow_last = float(combos_map[last_cid]["flow"])
    if flow_last <= 0:
        # 不合理，直接返回
        sol["start_time"] = sol["start_time"].dt.strftime("%Y-%m-%dT%H:%M:%S.%f").str.rstrip("0").str.rstrip(".")
        sol["end_time"]   = sol["end_time"].dt.strftime("%Y-%m-%dT%H:%M:%S.%f").str.rstrip("0").str.rstrip(".")
        return sol

    # 需要调整的时长（小时）
    need_hours = delta / flow_last  # 正表示需要延长，负表示需要缩短

    if need_hours > 0:
        # 尝试向后延长，不能超过当天24:00
        max_extend_hours = max(0.0, (day_end - last_end).total_seconds() / 3600.0)
        dt_hours = min(need_hours, max_extend_hours)
        new_end = last_end + pd.Timedelta(hours=dt_hours)
        sol.loc[last_idx, "end_time"] = new_end
    else:
        # 缩短，不能早于 start_time
        max_shrink_hours = max(0.0, (last_end - last_start).total_seconds() / 3600.0)
        dt_hours = min(abs(need_hours), max_shrink_hours)
        new_end = last_end - pd.Timedelta(hours=dt_hours)
        sol.loc[last_idx, "end_time"] = new_end

    # （可选）如果还不够精确，可再次微调：这里再算一次并二次微调（通常一次就够）
    current_volume2 = _compute_total_volume(sol, combos_map)
    delta2 = required_volume - current_volume2
    if abs(delta2) > tol:
        # 再做一次很小的“点修”（<= 1 分钟）以提高命中率
        tiny_hours = max(min(abs(delta2 / flow_last), 1.0/60.0), 0.0)  # ≤ 1 分钟
        if delta2 > 0:
            # 尝试再向后 +tiny
            if sol.loc[last_idx, "end_time"] + pd.Timedelta(hours=tiny_hours) <= day_end:
                sol.loc[last_idx, "end_time"] += pd.Timedelta(hours=tiny_hours)
        else:
            # 再向前 -tiny
            if sol.loc[last_idx, "end_time"] - pd.Timedelta(hours=tiny_hours) >= last_start:
                sol.loc[last_idx, "end_time"] -= pd.Timedelta(hours=tiny_hours)

    # 转回 isoformat 字符串（保留毫秒/微秒但去掉多余0）
    sol["start_time"] = sol["start_time"].dt.strftime("%Y-%m-%dT%H:%M:%S.%f").str.rstrip("0").str.rstrip(".")
    sol["end_time"]   = sol["end_time"].dt.strftime("%Y-%m-%dT%H:%M:%S.%f").str.rstrip("0").str.rstrip(".")

    return sol
```
